## <u>Detecting Outliers</u>

```python
fig, axs = plt.subplots(3, 3, figsize=(8,10))
fig.tight_layout(pad=4.0)
features = ['Open', 'High', 'Low', 'Previous Close', 'Last Traded Price','Share Volume', 'Value (Indian Rupee)', '52 Week High', '52 Week Low']
for f,ax in zip(features,axs.ravel()):
    ax=sns.boxplot(ax=ax,data=df,y=df[f])
    ax.set_title('Feature:'+ f)
```

# <u>Data Visualization</u>

### <u>Distributions of Industries</u>

```python
plt.figure(figsize=(30,10))
sns.countplot(x ='Industry', data = df)
plt.xticks(rotation=90)
```

## **<u>Observations</u>**

* Financial Services Industry contributes more value counts
* Forest Materials and Diversified industries have low value counts

```python
# a graphical representation of data using colors to visualize the value of the matrix.
sns.heatmap(df.corr(),annot=True,linewidths=0.2) #Brighter color represent less common or activity values.
```

## <u>Mean and median of (Open, High, Low, Previous Close, Last Traded Price) for each Industry</u>

```python
# Using groupby() method we can split the dataset into subsets to make computations easier.We can calculate median using the median() method.
df.groupby(['Industry']).agg({ 'Open':[np.mean, np.median],'High':[np.mean, np.median],'Low':[np.mean, np.median],'Previous Close':[np.mean, np.median],'Last Traded Price':[np.mean, np.median]})
```

## <u>Average Valuation of each Industry</u>

```python
df.groupby(['Industry']).agg({ 'Value (Indian Rupee)':[np.mean]})
```

```python
df.groupby(['Industry']).agg({ 'Value (Indian Rupee)':[np.mean]}).plot(kind="bar",figsize=(10,10))
```

## <u>Observations</u>
- Oil Gas and Consumable Fules Industry has the highest valuation followed by Metals & Publication Industry.
- Diversified Industry has lowest valuations followed by Textile Industry.

## <u>Scatter plot for Valuations</u>


```python
df.plot.scatter(x='Industry', y='Value (Indian Rupee)')
plt.xticks(rotation=90)            # xticks function is used to get or set the current tick locations and labels of the x-axis. 
```

```python
df.groupby(['Industry']).agg({ 'Share Volume':[np.mean]}).plot(kind="pie",subplots=True,figsize=(10,10))
```

## <u>Observations</u>
- Oil Gas & Consumable Fuels industries has the highest share volume.
